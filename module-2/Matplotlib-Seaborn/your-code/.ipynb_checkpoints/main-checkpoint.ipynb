#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.linspace(0, 5, 100)
y1 = x**2
y2 = np.exp(x)
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, sharex=False)
ax1.plot(x, y1)
ax1.plot(x, y2)
#ax1.legend(x**2, exp(x))

ax2.set_yscale('log')
ax2.plot(x, y1)
ax2.plot(x, y2)
#ax2.legend(x**2, exp(x))

#elles se comportent de la même manière mais sont représentées différemment

#sharex=False : each subplot x- or y-axis will be independent.
```

#### Plot the correlation matrix using seaborn.

```python
# Compute the correlation matrix
corr = titanic.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=.5, vmin=.6, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.heatmap(corr, annot=True)
```
